# **Sem aumentação**

```python
# Carregar os dados originais utilizando os datasets ainda não aumentados
data = pd.read_csv('/content/drive/MyDrive/dataset/tweetfake.csv', sep=',')

# Função de pré-processamento
def preprocess_text(text):
    # Converte para minúsculas
    text = text.lower()
    # Remove pontuações
    text = ''.join([char for char in text if char not in string.punctuation])
    # Tokenização
    tokens = word_tokenize(text)
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('portuguese')]
    return ' '.join(tokens)

data['text'] = data['text'].apply(preprocess_text)

# Definir X e y
X = data['text']
y = data['classificacao']

# Dividir os dados em treino e teste (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Pré-processamento de texto e vetorização
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train.astype('U'))
X_test_tfidf = tfidf_vectorizer.transform(X_test.astype('U'))

# Inicializar os modelos
rf_model = RandomForestClassifier()
knn_model = KNeighborsClassifier()
dt_model = DecisionTreeClassifier()
mlp_model = MLPClassifier()

# Treinar os modelos
rf_model.fit(X_train_tfidf, y_train)
knn_model.fit(X_train_tfidf, y_train)
dt_model.fit(X_train_tfidf, y_train)
mlp_model.fit(X_train_tfidf, y_train)

# Avaliação dos modelos com cross_validate
scoring = ['accuracy']
rf_scores = cross_validate(rf_model, X_train_tfidf, y_train, cv=5, scoring=scoring)
knn_scores = cross_validate(knn_model, X_train_tfidf, y_train, cv=5, scoring=scoring)
dt_scores = cross_validate(dt_model, X_train_tfidf, y_train, cv=5, scoring=scoring)
mlp_scores = cross_validate(mlp_model, X_train_tfidf, y_train, cv=5, scoring=scoring)

# Avaliar modelos com dados de teste
rf_test_accuracy = accuracy_score(y_test, rf_model.predict(X_test_tfidf))
rf_test_recall = recall_score(y_test, rf_model.predict(X_test_tfidf), pos_label='fake')
rf_test_precision = precision_score(y_test, rf_model.predict(X_test_tfidf), pos_label='fake')
rf_test_f1 = f1_score(y_test, rf_model.predict(X_test_tfidf), pos_label='fake')

knn_test_accuracy = accuracy_score(y_test, knn_model.predict(X_test_tfidf))
knn_test_recall = recall_score(y_test, knn_model.predict(X_test_tfidf), pos_label='fake')
knn_test_precision = precision_score(y_test, knn_model.predict(X_test_tfidf), pos_label='fake')
knn_test_f1 = f1_score(y_test, knn_model.predict(X_test_tfidf), pos_label='fake')

dt_test_accuracy = accuracy_score(y_test, dt_model.predict(X_test_tfidf))
dt_test_recall = recall_score(y_test, dt_model.predict(X_test_tfidf), pos_label='fake')
dt_test_precision = precision_score(y_test, dt_model.predict(X_test_tfidf), pos_label='fake')
dt_test_f1 = f1_score(y_test, dt_model.predict(X_test_tfidf), pos_label='fake')

mlp_test_accuracy = accuracy_score(y_test, mlp_model.predict(X_test_tfidf))
mlp_test_recall = recall_score(y_test, mlp_model.predict(X_test_tfidf), pos_label='fake')
mlp_test_precision = precision_score(y_test, mlp_model.predict(X_test_tfidf), pos_label='fake')
mlp_test_f1 = f1_score(y_test, mlp_model.predict(X_test_tfidf), pos_label='fake')
```

# **Função para augmentação**

```python
import pandas as pd
from googletrans import Translator
import random
import time

def backtranslate(text, dest_lang='en', src_lang='pt', chunk_size=5000):
    translator = Translator()

    # Dividir o texto em partes de tamanho chunk_size
    parts = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

    # Traduzir e concatenar as partes traduzidas
    translated_parts = []
    total_parts = len(parts)
    for i, part in enumerate(parts):
        try:
            # Traduzir a parte para o idioma de destino
            translated_part = translator.translate(part, src=src_lang, dest=dest_lang).text
            time.sleep(random.uniform(1, 3))
            # Traduzir a parte de volta para o idioma original
            backtranslated_part = translator.translate(translated_part, src=dest_lang, dest=src_lang).text
            # Adicionar a parte traduzida
            translated_parts.append(backtranslated_part)
            print(f"Backtranslating part {i+1}/{total_parts}")
        except Exception as e:
            # Em caso de erro, manter a parte original
            print(f"Error translating part {i+1}/{total_parts}: {e}")
            translated_parts.append(part)

    # Juntar as partes traduzidas de volta em uma string
    backtranslated_text = ' '.join(translated_parts)

    return backtranslated_text

# Carregar o dataset
dataset = pd.read_csv('synonyms_augmented.csv')

# Obter o total de linhas no dataset
total_lines = len(dataset)

# Aplicar backtranslation à coluna de texto do dataset
for idx, row in dataset.iterrows():
    augmented_text = row['augmented_text']
    print(f"Backtranslating row {idx+1}/{total_lines}")
    dataset.at[idx, 'texto_backtranslated'] = backtranslate(augmented_text)

# Salvar o novo dataset em um arquivo CSV separado
dataset.to_csv('back_translate_augmented.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import nltk
import random
import time  # Adicionando import do módulo time
from data_augmentation_GASPLN import text_augmentation

nltk.download('stopwords')
nltk.download('rslp')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('omw-1.4')

data = pd.read_csv('synonyms_augmented.csv', sep=',')

def process_dataset_and_augment_text(dataset):
    # Lista para armazenar os textos aumentados formatados
    augmented_texts = []

    # Total de linhas no dataset
    total_lines = len(dataset)

    # Itera sobre as linhas do dataframe
    for index, row in dataset.iterrows():
        # Obtém a string da coluna 'preprocessed_news'
        text = row['augmented_text']

        # Divide o texto em chunks de no máximo 5000 caracteres
        chunks = [text[i:i+5000] for i in range(0, len(text), 5000)]

        # Inicializa a variável para armazenar o texto aumentado
        augmented_text = ""

        # Itera sobre os chunks do texto
        for chunk in chunks:
            try:
                # Chama a função de aumentação de texto com os parâmetros desejados para cada chunk
                augmented_chunk = text_augmentation(chunk,
                                                    use_synonyms=True, synonyms_percentage=0.4,
                                                    use_back_translation=True, languages=['en', 'es'], translator='google',
                                                    use_character_swap=False, char_swap_prob=0.3,
                                                    use_random_swap=False, num_words=2, random_swap_prob=0.2,
                                                    use_add_noise=False, word_noise_prob=0.3, char_noise_prob=0.3)
                augmented_text += augmented_chunk
            except Exception as e:
                # Em caso de erro na tradução, mantém o chunk original
                augmented_text += chunk
                print(f"Error processing chunk {index + 1}/{total_lines}: {e}")

            # Adiciona um atraso aleatório de 1 a 3 segundos entre cada tradução
            time.sleep(random.uniform(1, 3))

        # Printa o índice atual em relação ao total de linhas
        print(f"Processing {index + 1}/{total_lines}")

        # Adiciona os textos originais, aumentados e as previsões à lista
        augmented_texts.append({'text': text, 'augmented_text': augmented_text, 'label': row['label']})

    # Cria um novo DataFrame com os resultados
    result_df = pd.DataFrame(augmented_texts)

    # Converte o DataFrame para uma string CSV
    result_csv = result_df.to_csv('synonyms_back_translate_full.csv', index=False, sep=',', quoting=1, doublequote=True, escapechar='\\')

    return result_csv

# Aplicar aumentação de dados
augmented_data = process_dataset_and_augment_text(data)
```
